```python
from pymatgen.ext.matproj import MPRester

with MPRester("5Gk9CNnZ5XHAgFnn") as m:
    # Structure for material id
    structure = m.get_structure_by_material_id("mp-11742")
```

```python
from pymatgen.analysis.diffraction.xrd import XRDCalculator as xrd

analyzer=SpacegroupAnalyzer(structure).get_primitive_standard_structure()
pattern=xrd('CuKa1').get_pattern(analyzer)

n=1
i=0
while i<3:
    print("Pic du diffractogramme n°",n)
    print("Intensité =", pattern.y[i])
    print("2\u03B8(°) =", pattern.x[i],"°")
    print("Indices hkl associés :",pattern.hkls[i][0]['hkl'])
    print("\n")
    n+=1
    i+=1
    
xrd('CuKa1').show_plot(analyzer)
```
