```python
from pymatgen.ext.matproj import MPRester

with MPRester("5Gk9CNnZ5XHAgFnn") as m:

    # Structure for material id
    structure = m.get_structure_by_material_id("mp-11742")

    # Dos for material id
    dos = m.get_dos_by_material_id("mp-11742")

    # Bandstructure for material id
    bandstructure = m.get_bandstructure_by_material_id("mp-11742")
```

## Inversion de centre (0,0,0) de l'atome de Cs en (0,0,0.5)

```python
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer as sga
from pymatgen.core.operations import SymmOp as sym


primStru=sga(structure).get_primitive_standard_structure()
convStru=sga(sga(primStru).get_conventional_standard_structure())
symOp=convStru.get_symmetry_operations()

op1=symOp[1]
print("Coordonnées après transformation :")
print(op1.operate((0,0,0.5)))
print("Matrice de rotation :")
print(op1.rotation_matrix)
print("Vecteur de translation :")
print(op1.translation_vector)
print()
```
